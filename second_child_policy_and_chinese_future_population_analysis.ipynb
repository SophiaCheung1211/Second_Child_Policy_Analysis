# **Predict Population from 2018 to 2030 (Total)**

```
def total_population_function(x,col_index): #col_index is the index to fill in total_pop_2017_2030 data frame
  # equation 1
  import ast
  #calculation of Male
  for i in range(1,91):
    value = (total_pop_2017_2030.iloc[i-1,col_index-3])*(1-(mr.iloc[i,x-2000+1]*1.5/1.3/1000)) + nm.iloc[i,x-1990]*(0.5)
    total_pop_2017_2030.iloc[i,col_index] = value

  #calculation of Female
  for i in range(1,91):
    value = (total_pop_2017_2030.iloc[i-1,col_index-2])*(1-(mr.iloc[i,x-2000+1]/1.3/1000)) + nm.iloc[i,x-1990]*(0.5)
    total_pop_2017_2030.iloc[i,col_index+1] = value

  #filling in zero in total_pop_2017_2030 for zero age
  total_pop_2017_2030.iloc[0,col_index] = 0
  total_pop_2017_2030.iloc[0,col_index+1] = 0

  #predict total_pop_2017_2030 for x_Total (actual total, not male + female)
  for i in range(1,91):
    value = (total_pop_2017_2030.iloc[i-1,col_index-1])*(1-mr.iloc[i,x-2000+1]/1000) + nm.iloc[i,x-1990]
    total_pop_2017_2030.iloc[i,col_index+2] = value

  #---------------------------------------------------------------------------------------------------------------------------------  
  #predict newborn sex ratio
  newborn_sex_ratio = math.exp(14.96445-0.0050649*x)
  if newborn_sex_ratio > 102:
    newborn_sex_ratio = math.exp(14.96445-0.0050649*x)
  else:
    newborn_sex_ratio = 102
  
  #predict newborn population
  newborn = 0
  for i in range(15,50): #15 to 49
    female = total_pop_2017_2030.iloc[i,col_index-2]
    fertility_rate = (fertility.iloc[i-15,x-1990+1]/1000 ) #column index is 17 for year 2017 in fertility
    newborn += (female*fertility_rate)
  mortality_rate = (mr.iloc[0,x-2000+1]/1000) #column index is 18 for year 2017 in mr
  newborn *= (1-mortality_rate)

  #calculate newborn boys and girls
  newborn_M = newborn_sex_ratio/(newborn_sex_ratio+100)*newborn
  newborn_F = (1-(newborn_sex_ratio/(newborn_sex_ratio+100)))*newborn

  #put into the total data frame
  total_pop_2017_2030.iloc[0,col_index] = newborn_M
  total_pop_2017_2030.iloc[0,col_index+1] = newborn_F
  total_pop_2017_2030.iloc[0,col_index+2] = newborn_M + newborn_F
```

```
column_index = 3 #column index for 2018_M is 3 for total_pop_2017_2030
for i in range (2018,2051):
  total_population_function(i,column_index)
  column_index += 3 #increase to x_M 
```
